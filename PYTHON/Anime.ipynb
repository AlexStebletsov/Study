```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

filename = 'anime.csv'
default_missing = pd._libs.parsers.STR_NA_VALUES
default_missing.add('-')
loaded_df = pd.read_csv(filename, thousands=',',na_values=default_missing)
anime = loaded_df.copy()
s = pd.Series(anime['Episodes'])
anime['Episodes'] = pd.to_numeric(s, errors='coerce')
anime.head(10)
```

```python
t= anime.production.value_counts().sort_values()

counts = []
names = []
for name,count in t.items():
    if len(counts)>0 and counts[-1] == count:
        names[len(counts) -1]+=', '+str(name)
    else:
        names.append(str(name))
        counts.append(count)

print('Больше всего аниме выпустила  ', names[-1], )
plt.xticks(rotation=-90)
plt.bar(names,counts)
```

```python
t= anime.episodes.value_counts().sort_values()
counts = []
names = []
for name,count in t.items():
    if len(counts)>0 and counts[-1] == count:
        names[len(counts) -1]+=', '+str(name)
    else:
        names.append(str(name))
        counts.append(count)



plt.xticks(rotation=-90)
plt.bar(names,counts)
```

```python
t= anime.source.value_counts().sort_values()
counts = []
names = []
for name,count in t.items():
    if len(counts)>0 and counts[-1] == count:
        names[len(counts) -1]+=', '+str(name)
    else:
        names.append(str(name))
        counts.append(count)

        
print('Самый популярный источник: ', names[-1] )
plt.xticks(rotation=-90)
plt.bar(names[-3:],counts[-3:])
```

```python
t = anime.theme.value_counts().sort_values()
counts = []
names = []
for name,count in t.items():
    if len(counts)>0 and counts[-1] == count:
        names[len(counts) -1]+=', '+str(name)
    else:
        names.append(str(name))
        counts.append(count)

print('Самая популярная тема', names[-1])
plt.xticks(rotation=-90)
plt.bar(names,counts)
```

```python
t = anime.groupby('production').mean()['rating'].sort_values()
counts = []
names = []
for name,count in t.items():
    if len(counts)>0 and counts[-1] == count:
        names[len(counts) -1]+=', '+str(name)
    else:
        names.append(str(name))
        counts.append(count)

print('Три компании выпускающие лучшие анниме', names[-3:] )
plt.xticks(rotation=-90)
plt.bar(names,counts)
```
